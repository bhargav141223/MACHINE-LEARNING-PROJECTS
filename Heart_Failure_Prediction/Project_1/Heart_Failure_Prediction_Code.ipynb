```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.ensemble import VotingClassifier
```

```python
# Define the feature columns and the target column
feature_columns = [
     'age', 'anaemia','creatinine_phosphokinase', 'diabetes', 'ejection_fraction',
    'high_blood_pressure', 'platelets', 'serum_creatinine', 'serum_sodium',
    'sex', 'smoking', 'time']
target_column = 'DEATH_EVENT'
```

```python
# Read the CSV file
df = pd.read_csv('/content/heart_failure_clinical_records_dataset.csv')

# Separate the features and the target variable
x = df[feature_columns].values
y = df[target_column].values

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.2)

# Standardize the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Define classifiers
classifiers = {
    'Random Forest': RandomForestClassifier(n_estimators=600, max_depth=30, random_state=1),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=600, max_depth=30, random_state=1),
    'AdaBoost': AdaBoostClassifier(n_estimators=600, random_state=1),
    'Decision Tree': DecisionTreeClassifier(max_depth=30, random_state=1),
    'Support Vector Classifier': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Naive Bayes': GaussianNB(),
    'MLP Classifier': MLPClassifier(max_iter=1000, random_state=1),
    'LDA': LinearDiscriminantAnalysis(),
    'Gaussian Process': GaussianProcessClassifier(random_state=1),
    'Voting Classifier': VotingClassifier(estimators=[
        ('rf', RandomForestClassifier(n_estimators=600, max_depth=30, random_state=1)),
        ('gb', GradientBoostingClassifier(n_estimators=600, max_depth=30, random_state=1)),
        ('svc', SVC(probability=True))
    ], voting='soft')
}

# Function to train and evaluate a classifier
def train_and_evaluate(clf, clf_name):
    clf.fit(x_train_scaled, y_train)
    predictions = clf.predict(x_test_scaled)
    accuracy = accuracy_score(y_test, predictions)
    print(f"{clf_name} Success Rate: {accuracy * 100:.2f}%")

    cm = confusion_matrix(y_test, predictions)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'{clf_name} Confusion Matrix')
    plt.show()

    print("Confusion Matrix:")
    print(cm)
    print("\nClassification Report:")
    print(classification_report(y_test, predictions))

# Train and evaluate each classifier
for clf_name, clf in classifiers.items():
    train_and_evaluate(clf, clf_name)
```

# SMOTE + ENN

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.ensemble import VotingClassifier

from imblearn.combine import SMOTEENN
from collections import Counter

# Read the CSV file
df = pd.read_csv('/content/heart_failure_clinical_records_dataset.csv')

# Assuming the last column is the target variable and the rest are features
x = df[feature_columns].values
y = df[target_column].values

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.2)

# Standardize the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Apply SMOTE + ENN
sme = SMOTEENN(random_state=0)
x_train_resampled, y_train_resampled = sme.fit_resample(x_train_scaled, y_train)

print("Original training set shape:", Counter(y_train))
print("Resampled training set shape:", Counter(y_train_resampled))

# Define classifiers
classifiers = {
    'Random Forest': RandomForestClassifier(n_estimators=600, max_depth=30, random_state=1),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=600, max_depth=30, random_state=1),
    'AdaBoost': AdaBoostClassifier(n_estimators=600, random_state=1),
    'Decision Tree': DecisionTreeClassifier(max_depth=30, random_state=1),
    'Support Vector Classifier': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Naive Bayes': GaussianNB(),
    'MLP Classifier': MLPClassifier(max_iter=1000, random_state=1),
    'LDA': LinearDiscriminantAnalysis(),
    'Gaussian Process': GaussianProcessClassifier(random_state=1),
    'Voting Classifier': VotingClassifier(estimators=[
        ('rf', RandomForestClassifier(n_estimators=600, max_depth=30, random_state=1)),
        ('gb', GradientBoostingClassifier(n_estimators=600, max_depth=30, random_state=1)),
        ('svc', SVC(probability=True))
    ], voting='soft')
}

# Function to train and evaluate a classifier
def train_and_evaluate(clf, clf_name):
    clf.fit(x_train_resampled, y_train_resampled)
    predictions = clf.predict(x_test_scaled)
    accuracy = accuracy_score(y_test, predictions)
    print(f"{clf_name} Success Rate: {accuracy * 100:.2f}%")

    cm = confusion_matrix(y_test, predictions)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'{clf_name} Confusion Matrix')
    plt.show()

    print("Confusion Matrix:")
    print(cm)
    print("\nClassification Report:")
    print(classification_report(y_test, predictions))

# Train and evaluate each classifier
for clf_name, clf in classifiers.items():
    train_and_evaluate(clf, clf_name)
```
